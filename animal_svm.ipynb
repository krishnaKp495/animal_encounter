```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("animals.csv")
print(df.head(1))
```

```python
x = df.drop(columns = 'winner', axis = 1)
x = x.dropna(axis=1, how='all')
```

```python
print(x)
```

```python
y = df['winner']
print(y)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
standard_data = scaler.transform(x)     # this can be done together with fit using fit_transform

print(standard_data)
```

```python
x = standard_data      # standardized data again stored in x
y = df['winner']
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size = 0.2, stratify = y, random_state = 42)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
classifier = svm.SVC(kernel= 'linear')
```

```python
classifier.fit(x_train, y_train)
```

```python
# accuracy score on the training data

x_train_prediction = classifier.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy score of the training data:', training_data_accuracy )
```

```python
# accuracy score on the test data

x_test_predict = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_predict, y_test)
print('Accuracy score of the test data:',test_data_accuracy)
```

```python
def animalvs(input_data):
    
    
    # changing the input data to numpy array
    
    input_data_as_numpy = np.asarray(input_data)
    
    # reshape the array as we are predicting for one instance
    
    input_data_reshaped = input_data_as_numpy.reshape(1,-1)
    
    # standardize the array
    
    std_data = scaler.transform(input_data_reshaped)
    prediction = classifier.predict(std_data)
    
    
    if (prediction[0] == 0):
        print(" the first animal is winner")
    else:
        print("the second animal is winner")
```

```python
input_data = (1200, 4.5, 35, 6.5, 200, 40, 8, 9, 7,	500, 4.5, 35, 9.5, 4400, 50, 9, 7.5, 8)
animalvs(input_data)
```
